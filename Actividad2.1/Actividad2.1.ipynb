```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Porto_Portugal.csv')
df = df.drop(columns=["Unnamed: 0"])
```

```python
df
```

**a) 5 "Host name" diferentes**

```python
filtro1 = df[(df['host_name'] == 'Naomi') | (df['host_name'] == 'Camila') | (df['host_name'] == 'Ann-Marie') | (df['host_name'] == 'Jo') | (df['host_name'] == 'Tricia')]
filtro1
```

```python
filtro1.to_csv("host_name.csv")
```

**b) Hosts que se hayan unido a Airbnb "después del año 2020"**

```python
filtro2 = df[(df['host_since'] >= '2023-01-01')]
filtro2
```

```python
filtro2.to_csv("host_since2020.csv")
```

**c) Hosts que responden "a más tardar 1 día"**

```python
filtro3 = df[(df['host_response_time'] == 'within a day')]
filtro3
```

```python
filtro3.to_csv('host_response_time.csv')
```

**d) Registros de filas que "Múltiplos de 200... (400,600,800, etc.)**

```python
filtro4 = df.iloc[(range(200, 6201, 200))]
filtro4
```

```python
filtro4.to_csv('multiplos200.csv')
```

**e) Registros de los tipos de cuarto: "Entire home" y "Private room"**

```python
filtro5 = df[(df['room_type'] == 'Entire home/apt') | (df['room_type'] == 'Private room')]
filtro5
```

```python
filtro5.to_csv("room_type.csv")
```

**f) Registros que "a lo mucho cuenten con 4 camas"**

```python
filtro6 = df[(df['beds'] <= 4)]
filtro6
```

```python
filtro6.to_csv('beds.csv')
```

**g) Los registros que "sean superhost"**

```python
filtro7 = df[(df['host_is_superhost'] == 't')]
filtro7
```

```python
filtro7.to_csv('is_superhost.csv')
```

**h)Registros de columnas impares** 

```python
filtro8 = df.iloc[:, 1::2]
filtro8
```

```python
filtro8.to_csv('impares.csv')
```
