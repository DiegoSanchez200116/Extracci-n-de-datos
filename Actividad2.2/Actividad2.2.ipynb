```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Porto_Portugal.csv')
df = df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
df.info()
```

a) Los registros de los host cuyo “host_response_rate” sea mayor a 95% y que no sean “superhost”

```python
df['host_response_rate']=df['host_response_rate'].str.replace('%','')
df['host_response_rate']=df['host_response_rate'].astype(float)
```

```python
filtro1 = df[(df['host_response_rate'] >= 95) & (df['host_is_superhost'] == 'f')]
filtro1
```

```python
filtro1.to_csv("host_response_rate.csv")
```

b) Los registros cuyas propiedades no tengan “1 baño”

```python
filtro2 = df[(df['bathrooms'] != 1)]
filtro2
```

```python
filtro2.to_csv("bathrooms.csv")
```

c) Los registros que tengan un precio entre “100 y 200 US” (Realizar conversión de moneda)

```python
df['price']=df['price'].str.replace('$','')
df['price']=df['price'].str.replace(',','')
df['price']=df['price'].astype(float)
```

```python
#1.05 dólares = 1 euro
filtro3 = df[(df['price'] >= 95.05) & (df['price'] <= 190.1)]
filtro3
```

```python
filtro3.to_csv("price.csv")
```

d) Los registros que sean “instant_bookable” y que hallan verificado su perfil “host_identity_verified”

```python
filtro4 = df[(df['instant_bookable'] == 't') & (df['host_identity_verified'] == 't')]
filtro4
```

```python
filtro4.to_csv("verified.csv")
```

e) Los registros de los tipos de cuarto: “hotel room” o “Private room”

```python
filtro5 = df[(df['room_type'] == 'Entire home/apt') | (df['room_type'] == 'Private room')]
filtro5
```

```python
filtro5.to_csv("room_type.csv")
```

f) Los registros que tengan mas de 4 “accommodates”

```python
filtro6 = df[(df['accommodates'] > 4)]
filtro6
```

```python
filtro6.to_csv("accomodation.csv")
```
